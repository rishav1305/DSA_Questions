```python
import numpy as np

def check_repetitions(slist):
    return len(slist) == len(set(slist))

def validate_list(slist):
    print(slist)
    slist = [x for x in slist if x != '.']
    return check_repetitions(slist)

def validate_rows(input):
    for row in input:
        if not validate_list(row):
            return False
    return True
    
def create_box(s, e, input):
    output = []
    for i in range(3):
        for j in range(3):
            output.append(input[s+i][e+j])
    return output

def validate_boxes(input):
    for s in range(0, 7, 3):
        for e in range(0, 7, 3):
            slist = create_box(s, e, input)
            if not validate_list(slist):
                return False
    return True


def valid_sudoku(input):
    print('Vertical check')
    vertical = validate_rows(input) 
    print(f'Vertical check DONE - {vertical}')
    print('Horizontal check')
    horizontal = validate_rows(np.array(input).T.tolist()) 
    print(f'Horizontal check DONE - {horizontal}')
    print('Boxes check')
    boxes = validate_boxes(input)
    print(f'Boxes check DONE - {boxes}')
    return vertical and horizontal and boxes

valid_sudoku(testcases[1][0])
```
